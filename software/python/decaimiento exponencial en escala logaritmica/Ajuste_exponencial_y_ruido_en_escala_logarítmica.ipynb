```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_state = 42
np.random.seed(random_state)
plt.figure(figsize=(10,5))

for offset in [20, 10,0,-10, -20]:
    A = 200
    tau = 10
    tamruido = 1.2
    x = np.arange(0,100,0.01)
    ruido = tamruido * np.random.randn(len(x))
    y = offset + A*np.exp(-x/tau) + ruido

    plt.subplot(1, 2, 1)
    plt.scatter(x, y, s=1, label=f'offset = {offset}')
    plt.grid(True)
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')

    plt.subplot(1, 2, 2)
    plt.scatter(x, y, s=1)
    plt.yscale('log')
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')
```
